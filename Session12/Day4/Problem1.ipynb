```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams.update({'font.size': 18, 'font.weight' : 'bold'})
mpl.rcParams['axes.linewidth'] = 2
```

```python
df = pd.read_csv('Gaia_kepler_stellar_properties_tab2.csv' )
df.head
```

```python
cmap = cm.get_cmap('RdYlBu') 
fig, ax = plt.subplots(1, figsize=(8,8))

lower = np.log10(df['Teff']).min()
upper = np.log10(df['Teff']).max()
colors = plt.cm.RdYlBu((np.log10(df['Teff'])-lower)/(upper-lower))


# ax.scatter(np.log10(df['Teff']),df['logLsun'],marker ='.', alpha = .7, s = np.log(df['Rsun']+1)*80,
#            cmap = cmap, edgecolor = colors, facecolor= 'none')
ax.scatter(np.log10(df['Teff']),df['logLsun'],marker ='.', alpha = .7, s = np.log(df['Rsun']+1)*80,
           cmap = cmap, c = colors)

ax.set_xlim(4.35,3.35)
ax.set_xlabel('log T (K)')
ax.set_ylabel(r'log L (L$_\odot)$')

ax.tick_params('both', length=6, width=1.25, which='major',direction="in", pad = 6)
ax.tick_params('both', length=6, width=1.25, which='minor',direction="in" )
```
